```python
import numpy as np
import pandas as pd
import networkx as nx
import pickle
```

```python
# Load the graph from the file
with open('projected_graph_para.gpickle', 'rb') as f:
    G = pickle.load(f)
```

```python
for i in G.nodes(data=True):
    print(i)
```

```python
# Calculate the average degree of the graph
average_degree = sum(dict(G.degree()).values()) / G.number_of_nodes()

print(f"Average degree: {average_degree}")
```

```python
# Assuming G is your graph
connected_components = list(nx.connected_components(G))


# Print the connected components
for component in connected_components:
    print(len(component))
```

```python
len_CC = sorted([len(cc) for cc in connected_components], reverse=True)
```

```python
len_CC
```

```python
largest_component = max(nx.connected_components(G), key=len)
print("Largest connected component:", largest_component)
```

```python
G_largest_CC = G.subgraph(largest_component).copy()
```

```python
from networkx.algorithms.community import greedy_modularity_communities
G_community = nx.community.greedy_modularity_communities(G_largest_CC)
G_modularity = nx.community.modularity(G_largest_CC, G_community)
G_modularity
```

```python
import csv
edges = G_largest_CC.edges(data=True)  # This will include any edge attributes

# Write the edges to a CSV file
with open('graph_edges.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    
    # Write the header (you can customize the columns based on your graph's attributes)
    csvwriter.writerow(['Source', 'Target', 'Weight'])
    
    # Write each edge and its attributes (e.g., weight)
    for u, v, data in edges:
        weight = data.get('weight', 1)  # Default weight is 1 if not present
        csvwriter.writerow([u, v, weight])

print("Edges have been exported to 'graph_edges.csv'")
```
